```python
from ect.recording import Oximeter
import serial
import matplotlib.pyplot as plt
```

```python
ser = serial.Serial('COM4')
```

```python
# Open seral port for Oximeter
oxi = Oximeter(serial=ser, sfreq=75)
oxi.setup()
```

```python
oxi.read(duration=10)
```

```python
oxi.plot()
plt.savefig('recording.png', dpi=600)
```

```python
plt.plot(oxi.diff)
```
