# Signal quality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from systole import import_ppg
from systole.detection import oxi_peaks, missed_beat, artifact_removal, hrv_subspaces
from systole.hrv import rmssd
from adtk.data import validate_series
```

```python
# Load data
oxi = import_ppg('Noisy1')[0, :]

# Extract peak vectors
signal, peaks = oxi_peaks(oxi)
```

```python
plt.figure(figsize=(12, 4))
#plt.plot(np.diff(np.where(peaks)[0]))
plt.plot(oxi)
plt.xlim(11000, 15000)
plt.ylim(90, 110)
```

```python
from adtk.visualization import plot
rr = oxi
time = pd.to_datetime(np.cumsum(rr), unit='ms')
df = pd.DataFrame({'rr': rr}, index=time)
df = validate_series(df.rr)
```

```python
plot(df, ts_linewidth=1)
```

```python
from adtk.detector import VolatilityShiftAD
volatility_shift_ad = VolatilityShiftAD(c=6.0, side='positive', window=50)
anomalies = volatility_shift_ad.fit_detect(df)
plot(df, anomaly_pred=anomalies, ts_linewidth=1, ap_color='red');
```

```python
from adtk.detector import LevelShiftAD
level_shift_ad = LevelShiftAD(c=6.0, side='both', window=5)
anomalies = level_shift_ad.fit_detect(df)
plot(df, anomaly_pred=anomalies, ts_linewidth=1, ap_color='red');
```

```python
from adtk.transformer import DoubleRollingAggregate
s_transformed = DoubleRollingAggregate(
    agg="median",
    window=2,
    diff="l2").transform(df)

plot(s_transformed.rename("Diff rolling quantiles"), ts_linewidth=1);
```
